```python
# answer the question "What graduate level courses are offered in semester X?" for each semester, where the graduate level courses are indicated in GraduateLevel column and the course should indicate the courseCode, courseName, the number of credit hours, and the technical interest area (tia), if it is not nan. The number of credit hours is indicated in TotalHours column
semesters = ["Fall\n2023", "Spring\n2024", "Summer\n2024", "Fall\n2024", "Spring\n2025"]
semestersNames = ["Fall 2023", "Spring 2024", "Summer 2024", "Fall 2024", "Spring 2025"]

for i in range(5):
    st+=(f"<s>[INST]What graduate level courses are offered in {semestersNames[i]}?[/INST]")
    st+=(f"The graduate level courses offered in {semestersNames[i]} are:")
    for courseCode in df[(df[semesters[i]]=='X') & (df['GraduateLevel']=='Yes')]['CourseCode']:
        courseName = df[df['CourseCode']==courseCode]['CourseName'].values[0]
        tia = df[df['CourseCode']==courseCode]['TechnicalInterestArea'].values[0]
        if not pd.isnull(tia):
            st+=(f"{courseCode}:{courseName} ({df[df['CourseCode']==courseCode]['TotalHours'].values[0]} SCH), in {tia}")
        else:
            st+=(f"{courseCode}:{courseName} ({df[df['CourseCode']==courseCode]['TotalHours'].values[0]} SCH)")
    st+=("</s>")
    arrayOfq.append(st)
    st=""
    
    
```
